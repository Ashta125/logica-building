2. Pass / Fail Analyzer
 Problem Statement:
 Analyze student results.

```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count = pass_count + 1
    else:
        fail_count = fail_count + 1

print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

3. Simple Data Cleaner
Problem Statement:
Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]


```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []
```

Remove extra spaces and convert to lowercase

```python
for name in names:

    clean_name = name.strip().lower()
    cleaned_names.append(clean_name)

print("Cleaned Names List:", cleaned_names)
```

Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]


```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")
    
# Flag messages longer than 10 characters
    if length > 10:
        print("⚠️ This message is longer than 10 characters")
```

5. Error Message Detector
Problem Statement:
Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]


```python

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR messages:", error_count)
```
